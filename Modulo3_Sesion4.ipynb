# **Gráficas Matemáticas**

```python
# Importando librerías
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Genera el vector de elementos desde -10 hasta 10 con intervalo 1
x = np.arange(-10, 11, 0.1)
y = x**2
```

```python
fig = plt.plot(x, y)
```

```python
# Ahora vamos a darle estilo a la gráfica original
fig = plt.plot(x, y, color = 'red', linestyle = 'dashed')
plt.title("Función Cuadrática") # Titulo de la Figura
plt.xlabel("Tiempo") # Titulo del Eje X
plt.ylabel("Velocidad") # Titulo del Eje Y
plt.grid() # Muestra una malla atras de la gráfica
plt.show() # Quita el mensaje del "entorno figura" (probar sin este codigo)
```

```python
# Práctica
x = np.arange(-10, 11, 0.1)
y = np.log(x)
fig = plt.plot(x, y)
plt.title("Función Logaritmo")
plt.xlabel("Tiempo")
plt.ylabel("Mora")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,20,0.1)

y1 = 20 - x

y2 = 4* x ** 0.5

Salmon = (250/250, 128/250, 114/250)

IndianRed = (205/250, 92/250, 92/250)

plt.plot(x, y1, color=Salmon, label='lineal')
plt.plot(x, y2, color=IndianRed, label='cuadratica')
plt.title('Funciones')
plt.legend(loc='center right')
plt.show()

```

# **Subplots**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y1 = x**2
y2 = x**0.5
y3 = np.cos(x)
y4 = np.tan(x)

fig , axs = plt.subplots(2,2) #2 filas y 2 columnas

axs[0,0].plot(x, y1)
axs[0,0].set_xlabel('Tiemp')
axs[0,0].set_ylabel('Velocidad')
axs[0,0].set_title('Cuadrática', fontsize = '8')

axs[1,0].plot(x, y2)
axs[1,0].set_title('Raiz', fontsize = '8')

axs[0,1].plot(x, y3)
axs[0,1].set_title('Coseno', fontsize = '8')

axs[1,1].plot(x, y4)
axs[1,1].set_title('Tangente', fontsize = '8')

fig.suptitle('Funciones Matemáticas') #Le coloca titulo a todo el conjunto de gráficas
fig.tight_layout() # Evita que los titulos se superpongan a los ejes del otro
plt.show()
```

# **Histograma**

La librería **matplot** no es la mejor para trabajar histograma. Es mejor utilizar **Seaborn**

```python
x = [99, 77, 101, 89, 106, 74, 124, 78, 120, 120, 79, 59,90, 90]

fig, ax = plt.subplots()
ax.hist(x, bins = 6, edgecolor = 'white')
plt.title('Tamaño de Departamentos')
plt.show()
```

```python
x = [99, 77, 101, 89, 106, 74, 124, 78, 120, 120, 79, 59,90, 90, 200, 2]

plt.boxplot(x)
plt.show()

# En la gráfica se identifican el 200 y el 2 como outliers
```
