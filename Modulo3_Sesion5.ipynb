# **Ejercicio de Repaso**

```python
# Llamar a las librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generar el x
x = np.arange(-2, 2, 0.1)

# Generar f(y)
y = x**2 + 5*x +1

# Generar el Gráfico
plt.plot(x, y, color = 'Green')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Función Cuadrática')
plt.grid()
plt.show()
```

```python
x = np.arange(0, 4, 0.1)
y = 16 - x**2

plt.plot(x, y, label='f(x)', color = 'Yellow', linestyle = 'dashed')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Diego')
plt.legend()
plt.grid()
plt.show()
```

```python
# generar x1
x1 = np.arange(0,4,0.01)
x2 = np.arange(4,10,0.01)

f1 = -x1+3
f2 = x2-5

plt.plot(x1,f1,color='red')
plt.plot([4],[-0.98],marker ='o',color='red')
plt.plot(x2,f2,color='red')
plt.show()
```

```python
x = np.arange(0,10,0.01)
f = x
g = 10 - x

plt.plot(x, f, color = 'red', label = 'Funcion f(x)')
plt.plot(x, g, color = 'blue', label = 'Funcion g(x)')
plt.xlabel('Eje X')
plt.ylabel('Eje y')
plt.legend()
plt.grid()
plt.title('Ejercicio')
plt.show()
```

```python
x = np.arange(-1,5,0.01)
f = x**2
g = 5 - x

plt.plot(x, f, color = 'green', label = 'Funcion f(x)')
plt.plot(x, g, color = 'yellow', label = 'Funcion g(x)')
plt.xlabel('Eje X')
plt.ylabel('Eje y')
plt.legend()
plt.grid()
plt.title('Diego')
plt.show()
```
