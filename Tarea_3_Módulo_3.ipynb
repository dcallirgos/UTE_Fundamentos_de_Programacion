```python
# Generar el x
x = np.arange(-3, 3, 0.1)

# Generar las funciones dependientes
f = x**2
g = 9 -  x**2

# Generar el Gráfico
plt.plot(x, f, color = 'red', label = 'Funcion f(x)')
plt.plot(x, g, color = 'blue', label = 'Funcion g(x)')
plt.xlabel('Eje X')
plt.ylabel('Eje y')
plt.legend()
plt.grid()
plt.title('Funcioes Cuadráticas')
plt.show()
```

# 3. Gráficas Residudos Sólidos

```python
# Llamar a las librerías
import pandas as pd
```

```python
# Accediendo a los datos de Tarea-data.xlsx
df = pd.read_excel('Tarea-data.xlsx')
df.head()
```

Graficando con Pandas

```python
# Vamos a gráficar la media de residuos producidos por año
df1 = df.groupby(['Año'])['Residuos'].mean()
df1.head()
```

```python
ax = df1.plot.bar(['Residuos']);
for container in ax.containers:
  ax.bar_label(container)
```

```python
# Vamos a gráfica la suma de la población en los distritos por año
df2 = df.groupby(['Año'])['Poblacion'].sum()/1000
df2.head()
```

```python
ax = df2.plot.bar(['Poblacion']);
for container in ax.containers:
  ax.bar_label(container)
```

Comentario: De las dos gráficas anteriores vemos que existe una correlación entre el crecimiento de la población y la producción de residuos

```python
# Vamos a gráficar la suma de residuos producidos por cada distrito en los 4 años
df3 = df.groupby(['Distrito'])['Residuos'].sum()/1000
df3.head()
```

```python
ax = df3.plot.bar(['Residuos']);
for container in ax.containers:
  ax.bar_label(container)
```

De la gráfica anterior se evidencia que San Juan de Lurigancho es el distrito que más ha producido resiudos; lo cual es intuitivo respecto al volumen de su problación
