# **Limpieza y Transformación de Datos**

```python
# Importar las librerías
import pandas as pd
import numpy as np
```

```python
# Importar base de datos
data = pd.read_excel("Datos_covid-SIN LIMPIAR.xlsx")
data.head(30)
```

```python
# Opción 1: Eliminar datos vacios
data1 = data.dropna()
# Pero dropna elimina todas las celdas vacias incluso podría borrar la celda que pertenece a un registro
# que tenga celdas adyacentes a las vacias que si tienen datos
data1.head(30)
```

```python
# Opción 2: Eliminar datos vacios
data2 = data.dropna(how = 'all')
# En este caso elimina las filas completamente vacías
data2.head(30)
```

```python
# Opción 2: Eliminar datos vacíos
data3 = data.dropna(subset = ['edad', 'Peso', 'Altura'])
# En este caso elimina los vacios de las columna a estudiar
data3.head(30)
```

```python
# Eliminar filas Duplicados cuando todas las variables coinciden
data4 = data.drop_duplicates()
data4.head(30)
```

```python
data = pd.read_excel("Datos_covid-SIN LIMPIAR.xlsx")
data = data.drop_duplicates()
data = data.dropna(how = 'all')
data.head(20)
```

```python
data['UCI'] = np.where(data['flag_uci'] == 1, 'SI', 'NO')
data = data.drop(['flag_uci'], axis = 1)
data.head(20)
```

```python
# Crear una nueva variable que sea el IMC
data['IMC'] = data['Peso']/((data['Altura']/100)**2)
data.head(30)
```

# **Librería Plotly**

Esta librería permite generar gráficos con los cuales se puede interactuar

```python
# Importando librerías
import plotly.express as px
import pandas as pd
```

```python
# Importando datos
data = pd.read_excel("Data-alquiler.xlsx")
```

```python
data. head()
```

```python
# Gráfico de Dispersión
fig = px.scatter(data, x = 'superficie', y = 'alquiler')
fig.show()
```

```python
fig = px.scatter(data, x = 'superficie', y = 'alquiler', color = 'distrito')
fig.show()
```

```python
fig = px.scatter(data, x = 'superficie', y = 'alquiler', size = 'baños', color = 'baños')
fig.show()
```

```python
fig = px.scatter(data, x = 'superficie', y = 'alquiler', size = 'baños', color = 'baños', facet_row = 'año', facet_col = 'distrito', trendline = 'ols')
fig.show()
```

```python
fig = px.histogram(data, x = 'alquiler')
fig.show()
```

```python
fig = px.histogram(data, x = 'superficie', text_auto=True, nbins=30)
fig.show()
```

```python
fig = px.histogram(data, x = 'superficie', text_auto=True, nbins=20, facet_row = 'año', facet_col = 'distrito')
fig.show()
```

```python
fig = px.histogram(data, x = 'distrito', text_auto=True, facet_col = 'año')
fig.show()
```
