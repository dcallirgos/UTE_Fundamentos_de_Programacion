```python
# iloc sirve para especificar la aplicación de un cambio en un lugar específico
# En este caso se está reemplazando las F por 0 sólo en las columnas de la 1 a la 9 y en todas las filas
# y Está sobre escribiendo el resultado en la variable data_2022_1

data_2022_1.iloc[:, 1:9] = data_2022_1.iloc[:, 1:9].replace('F', 0)
```

```python
# Calculando el promedio de las Practicas C y agregandolo a nueva columna
practica_c = ['Practica C1',	'Practica C2',	'Practica C3',	'Practica C4']

def prom_pc(row):
    return (sum(row) - min(row)) / (len(row) - 1)

data_2022_1['prom_pc'] = data_2022_1[practica_c].apply(prom_pc, axis=1)
```

```python
# Calculando el promedio de las Practicas D y agregandolo a nueva columna
practica_d = ['Practica D1',	'Practica D2',	'Practica D3',	'Practica D4']

def prom_pd(row):
    return (sum(row) - min(row)) / (len(row) - 1)

data_2022_1['prom_pd'] = data_2022_1[practica_d].apply(prom_pd, axis=1)
```

```python
# Cambiando el Valor de Examen 3
# Si es F ==> 0
# Si es <> F == > Queda con su valor

examen = ['Examen 1',	'Examen 2',	'Examen 3']

data_2022_1['Examen 3'].replace('F', 0, inplace = True)
data_2022_1['Examen 2'].replace('N', 0, inplace = True)
```

```python
# Calculando el promedio de los Examenes y agregandolo a nueva columna

examen = ['Examen 1',	'Examen 2',	'Examen 3']

def prom_ex(row):
  examen1 = row['Examen 1']
  examen2 = row['Examen 2']
  examen3 = row['Examen 3']
  if (examen1 == 'F' and examen2 == 'F'):
    return 0
  elif examen1 == 'F':
    return (examen2 + examen3) / 2
  elif examen2 == 'F':
    return (examen1 + examen3) / 2
  else:
    return (examen1 + examen2) / 2

data_2022_1['prom_ex'] = data_2022_1.apply(prom_ex, axis = 1)
```

```python
def prom_final(row):
  prom_pc = row['prom_pc']
  prom_pd = row['prom_pd']
  prom_ex = row['prom_ex']
  return round(prom_pc*0.25 + prom_pd*0.15 + prom_ex*0.6)

data_2022_1['prom_final'] = data_2022_1.apply(prom_final, axis = 1)
```
